```python
# Importaci√≥n de fomulas
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
```

```python
sym.var('x y')
```

```python
def f(x,y):
    return (x-3)**2+(y-5)**2
```

```python
dfx = sym.diff(f(x,y), x)
dfy = sym.diff(f(x,y), y)
dfx, dfy
```

```python
xy_c = sym.solve([dfx, dfy],[x,y])
xy_c
```

```python
x_c = xy_c[x]
y_c = xy_c[y]
x_c, y_c
```

```python
d2fx = sym.diff(f(x,y),x,2)
d2fxy= sym.diff(f(x,y),x,y)
d2fy = sym.diff(f(x,y), y, 2)

H = sym.Matrix([[d2fx,d2fxy],[d2fxy,d2fy]]) #con mayusculas M
H
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

#Definir coord
xnum = np.linspace(-10, 10, 100)
ynum = np.linspace(-15, 15, 100)

X,Y = np.meshgrid(xnum, ynum)

ax.plot_surface(X,Y,f(X,Y))
ax.plot([x_c],[y_c],[f(x_c,y_c)], '*r', ms=15)
```
